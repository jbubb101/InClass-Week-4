```python
import networkx as nx
import matplotlib.pyplot as plt

# Create the directed graph and add edges
G = nx.DiGraph()
G.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'D')])

# Updated days associated with each edge (representing midpoint days)
days_map = {('A', 'B'): 5, ('B', 'C'): 3, ('C', 'D'): 7}

# Here are the colors tha I  chose
node_colors = ['red', 'orange', 'green', 'blue']

# These are the shapes that she wanted.
node_shapes = {'A': 'o', 'B': 'd', 'C': 'o', 'D': 's'}

# Edges and colors
red_edges = [('A', 'B'), ('B', 'C')]
edge_colours = ['black' if edge not in red_edges else 'red' for edge in G.edges()]
black_edges = [edge for edge in G.edges() if edge not in red_edges]

# Layout for the graph
pos = nx.spring_layout(G)

# Draw nodes with the specific shapes
for node, shape in node_shapes.items():
    nx.draw_networkx_nodes(G, pos, nodelist=[node], node_color=[node_colors.pop(0)], node_shape=shape, node_size=500)

# Draw labels for the nodes
nx.draw_networkx_labels(G, pos)

# Draw edges
nx.draw_networkx_edges(G, pos, edgelist=red_edges, edge_color='r', arrows=True)
nx.draw_networkx_edges(G, pos, edgelist=black_edges, arrows=False)

# Add midpoint nodes and labels for days between edges
mid_pos = {}  
mid_labels = {}  

for edge in G.edges():
    x_mid = (pos[edge[0]][0] + pos[edge[1]][0]) / 2
    y_mid = (pos[edge[0]][1] + pos[edge[1]][1]) / 2
    mid_pos[edge] = (x_mid, y_mid)

    mid_labels[edge] = f"{days_map[edge]} days"


nx.draw_networkx_labels(G, mid_pos, labels=mid_labels, font_color='blue')


plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a new graph
G = nx.Graph()

# Add nodes
G.add_nodes_from([1, 2, 3, 4, 5])

# Add some edges between the nodes
G.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 5), (4, 5)])

# Draw the network
pos = nx.spring_layout(G)  
nx.draw(G, pos, with_labels=True, node_color='pink', node_size=1000, font_size=15, font_weight='bold')

# Show the network diagram
```
