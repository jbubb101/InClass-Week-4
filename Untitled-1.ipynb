```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation Parameters
n_simulations = 10000




task_1 = {'mean':5,"st_dev":1}
task_2 = {'mean':10,"st_dev":2}
task_3 = {'mean':7,"st_dev":1}

task_1_sim= np.random.normal(task_1['mean'], task_1['st_dev'], n_simulations)
task_2_sim= np.random.normal(task_2['mean'], task_2['st_dev'], n_simulations)
task_3_sim= np.random.normal(task_3['mean'], task_3['st_dev'], n_simulations)

total_time_sim = task_1_sim + task_2_sim + task_3_sim

# Analyze the results
mean_total_time = np.mean(total_time_sim)
st_dev_total_time = np.std(total_time_sim)

print(f"Mean total time: {mean_total_time}")
print(f"Standard deviation of total time: {st_dev_total_time}")
```

```python
plt.figure(figsize=(12, 6))

plt.hist(task_1_sim, bins=10, color='green', edgecolor='black')
plt.title('Distribution of task 1')
plt.xlabel('Total Revenue (Millions USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))

plt.hist(task_2_sim, bins=10, color='indigo', edgecolor='black')
plt.title('Distribution of task 2')
plt.xlabel('Total Revenue (Millions USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))

plt.hist(task_3_sim, bins=10, color='purple', edgecolor='black')
plt.title('Distribution of task 1')
plt.xlabel('Total Revenue (Millions USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()


plt.figure(figsize=(12, 6))

plt.hist(total_time_sim, bins=10, color='orange', edgecolor='black')
plt.title('Distribution of total task time')
plt.xlabel('Total Revenue (Millions USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

print(f"Mean total time: {mean_total_time}")
print(f"Standard deviation of total time: {st_dev_total_time}")
print(f"Best Case scenario {total_time_sim.max()}")
print(f"Min Case scenario {total_time_sim.max()}")

```
